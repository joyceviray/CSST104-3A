# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# **Loading Dataset**

```python
data = pd.read_csv('Lesson_4_student_grades_dataset copy.csv')
data.columns
print(data.head)
```

# **Preparing Data for Modeling**

```python
X = data.drop('Grades', axis = 1)
Y = data['Grades']
```

# **Train and Split Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state= 42)
```

# **Fitting the Linear Regression**

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

# **Making Prediction**

```python
new_data = np.array([[5, 90, 80, 0, 30]])
Y_predict = model.predict(new_data)

grade = (Y_predict/60)*100

print(f"Predicted Grade:{grade[0]}")

```

```python
Y_predict = model.predict(X_test)
```

# **Making Squared Error**

```python
mse = mean_squared_error(Y_test, Y_predict)

print(mse)
print(Y_predict)
```

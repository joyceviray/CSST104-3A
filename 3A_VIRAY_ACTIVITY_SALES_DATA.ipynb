1. Importing Libraries

```python
import pandas as pd #Data Manipulation
import matplotlib.pyplot as plt # Data Visualization
```

2. Loading Datasets

```python
#Import sales_data.csv
data = pd.read_csv('sales_data_2.csv')
```

3. Data Loading and Initial Analysis

```python
statistics = data.describe()
```

4. Data Manipulation

```python
product_revenue = data.groupby('Product')['Revenue'].sum()
highest_revenue_product = product_revenue.idxmax()
average_quantity_per_day = data['Quantity Sold'].mean()

```

5. Data Visualization

```python
product_revenue.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue By Product')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

6. Analysis Report



```python
report = f"""

Analysis Report

1. Data Loadind and Initial Analysis:
  Basic statistic of the dataset:
  {statistics}

2. Data Manipulation
  Total revenue for each product:
  {product_revenue}

  Product with the highest revenue:
  Average quantity sold per day{average_quantity_per_day}

3. Data Visualization:
  Bar chart displaying revenue for each product is attached.

4. Insights:
  - The product {highest_revenue_product} generate the highest revenue.
  - On average, {average_quantity_per_day} units sold per day.



"""
```

```python
print(report)
```

7. Saving the report to a file

```python
with open('report.txt', 'w') as f:
    f.write(report)
```

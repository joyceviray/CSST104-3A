# **1. Import Libraries**

```python
 import pandas as pd
 import numpy as np
 import matplotlib.pyplot as plt
 from sklearn.linear_model import LinearRegression
```

# **2. Loading Dataset**

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.columns
print(data.head)
```

# **3. Preparing Data for Modelling**

```python
X = data[['Hours_of_Study']] #Independent Variable
Y = data[['Grades']] #Dependent Variable

```

# **4. Fitting the Linear Regression Model**

```python
model = LinearRegression()

model.fit(X,Y)
```

# **5. Getting Model Coefficient**

```python
slope = model.coef_[0]
intercept = model.intercept_

print('Slope:', slope)
print('Intercept:', intercept)
```

# **6. Visualizing the Linear Regression Line**

```python
plt.scatter(X,Y, color = 'orange')
plt.plot(X, model.predict(X),color = 'black')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title('Linear Regression of Grades to Hours of Study')
plt.show()

```

# **Making Predictions**

```python
new_X = [[40]]
predicted_Y = model.predict(new_X)

print('Value of X:', new_X)
print('Predicted Value of Y:', predicted_Y)
```
